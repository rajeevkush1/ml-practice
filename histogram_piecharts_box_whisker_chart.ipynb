# **PIE** **CHARTS** *italicized text*

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
# left.shape
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
# print(right)
# print (left)

labels=['right','left']
colours=['#abcedf','#ab374f']

plt.pie([right,left],labels=labels,colors=colours,autopct='%.2f %%')
plt.title("Preferred Foot")
plt.show()
```

pie chart -2

```python
fifa.Weight= [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]


# plt.style.use('ggplot')
#plt.style.use('seaborn')
plt.style.use('fivethirtyeight')
# plt.style.use('grayscale')

underweight = fifa.loc[fifa.Weight < 125].count()[0]
# print(underweight)
normal=fifa.loc[(fifa.Weight>=125) & (fifa.Weight<185)].count()[0]
#print(normal)
overweight=fifa.loc[fifa.Weight>=185].count()[0]
#print(overweight)
weights=[underweight,normal,overweight]

labels=['underweight','normal','overweight']
plt.figure(figsize=(10, 7))

explode=[.8,0,.4]    # EXPLODES THE PIE CHART...
plt.pie(weights,labels=labels,autopct='%.2f%%',pctdistance=.9,explode=explode)
plt.title("Weight Distribution",fontweight='bold',fontsize=30,color='blue')
plt.show()
```

# ***box-and-whisker chart***

```python
plt.style.use('default')

plt.figure(figsize=(5, 8))

barcelona = fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
realo=fifa.loc[fifa.Club=='Chelsea']['Overall']

labels=['FC Barcelona',"Real Madrid",'realo']

X=plt.boxplot([barcelona,madrid,realo],labels=labels   ,patch_artist=True , medianprops={'linewidth':2,'color':'blue'})  #patchartist property is used to allow "facecolour" function work
for box in X['boxes']:
  box.set(color='brown',linewidth=2)
  box.set(facecolor='#e0e0e8')

plt.title("Overall Distribution", fontsize=20, fontweight='bold',color='brown')
plt.ylabel('Player Overall')
plt.show()
```
