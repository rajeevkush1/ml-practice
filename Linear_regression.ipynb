# **LINEAR REGRESSION BASED QUESTION**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#from sklearn.metrics import accuracy_score
```

```python
#creating data
shard={'hours_study':[2,3,4,5,6,7,8,9,10],'Exam_score':[50,60,70,75,80,85,90,94,98]}  #Array dataframe creation
df=pd.DataFrame(shard,columns=['hours_study','Exam_score'])
print(df)
```

```python
#Feature extraction
x=df[['hours_study']]
y=df[['Exam_score']]
```

We train 80% data & test for 20% of data

```python
X_train, X_test,Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
#test_size = .2 (means 20%  of data is being tested)
# random_state is parameter is being set 42 for training same data kind
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
# user testing
user_input=float(input("enter no of hours you study"))
pridictin=(model.predict([[user_input]]))
```
