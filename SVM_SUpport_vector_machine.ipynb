```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
data={'age':[10,15,25,33,40,45,48,50,55,59], 'recharge':[149,155,160,161,169,180,181,185,190,200],'retain':[1,1,0,0,1,0,1,0,0,1]}
df=pd.DataFrame(data)
df
```

```python
x=df[['age','recharge']]
y=df['retain']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
SVC_model= SVC(kernel='linear',C=1)
SVC_model.fit(x_train,y_train)
y_pred=SVC_model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("accuracy:",accuracy)
```

CLAssification REport

```python
report= classification_report(y_test,y_pred)
print("classification report:\n",report)
```

```python
user_age=float(input('enter users age'))
user_recharge=float(input('enter users recharge'))
user_data=np.array([[user_age,user_recharge]])
prediction=SVC_model.predict(user_data)
if prediction==1:
  print('customer will retain')
else:
  print('customer will not retain')

#print(prediction)
```
