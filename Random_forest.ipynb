```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
 # Features array (Age, Salary, Gender)
X = [
    [25, 45000, 0],  # Male is encoded as 0
    [32, 52000, 1],  # Female is encoded as 1
    [41, 58000, 0],
    [19, 31000, 1],
    [28, 49000, 0],
    [35, 65000, 1],
    [23, 43000, 1],
    [45, 70000, 0]
]

# Target array (Buy or not)
y = [0, 1, 1, 0, 1, 1, 0, 1]
```

```python
data=pd.DataFrame(X,y)
print(data)
```

```python
X_test,X_train,y_test,y_train= train_test_split(X,y,test_size=.2,random_state=42)
```

```python
rf_classifier= RandomForestClassifier(n_estimators=100,random_state=42)
rf_classifier.fit(X_train,y_train)
```

```python
y_pred=rf_classifier.predict(X_test)
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
age=float(input('enter age'))
income = float (input ('enter income'))
gender= int(input('enter gender 0 for male 1 for fmale'))
user_input=np.array([[age,income,gender]])
user_pred=rf_classifier.predict(user_input)
if user_pred==0:
    print('not in mood to buy phone')
else:
    print('is ready to buy new phone')
```
