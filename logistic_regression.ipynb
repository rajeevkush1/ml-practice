```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x= np.array([[25,30, 1],[30,50, 1],[35,40, 0],[20,25, 0]])
y= np.array([1,1,0,0])
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
model= LogisticRegression()
model.fit(x_train, y_train)
#y_pred= model.predict(x_test)
accuracy= model.score(x_test, y_test)
print("Accuracy:", accuracy)
```

```python
user_age=float(input("enter users age"))
user_time_spent=float(input("enter users time spent"))
user_added_to_cart= float(input("enter 1 if added to card else 0 :"))
user_input = np.array([[user_age, user_time_spent, user_added_to_cart]])
prediction= model.predict(user_input)
if prediction[0]==1:
    print("user will buy the product")
else:
    print("user will not buy the product")
```
