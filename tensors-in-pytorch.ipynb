```python
import torch
import time
size=10000


matrix_cpu1=torch.rand(size,size)
matrix_cpu2=torch.rand(size,size)

start_time=time.time()
result=torch.matmul(matrix_cpu1,matrix_cpu2)
cpu_time=time.time() - start_time

print(f"cpu takes",cpu_time,"time for complete execution" )

matrix_gpu1=matrix_cpu1.to('cuda')
matrix_gpu2=matrix_cpu2.to('cuda')
start_time=time.time()
result=torch.matmul(matrix_gpu1,matrix_gpu2)
gpu_time=time.time() - start_time

print(f"gpu takes",gpu_time,"time for complete execution")

print("\nspeedup/ratio:",cpu_time/gpu_time)
```
