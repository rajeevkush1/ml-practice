```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x=np.array([[25,30000,1],[30,80000,4],[35,160000,5],[40,240000,2],[45,320000,3],[46,100000,4],[50,27000,3],[60,27000,1]] )
# 10=1,12=2,grad=3,pg=4,phd=5  level of study
y=np.array([0,0,1,1,1,1,0,0])
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
#user input
age=float(input('enter age'))
income = float (input ('enter income'))
graduation= int(input('enter graduation level'))
user_input=np.array([[age,income,graduation]])
user_pred=model.predict(user_input)
if user_pred==0:
    print('not in mood to buy phone')
else:
    print('is ready to buy new phone')
```
