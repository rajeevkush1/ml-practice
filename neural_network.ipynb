```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()  # used for changing exam_result into binary form from pass/fail
```

```python
#data for the model
study_time=[2,2.5,1,3,5,1.5,2,3.2,1.9]
marks_scored=[60,70,30,33,95,93,20,70,85]
exam_result=['pass','pass','fail','fail','pass','pass','fail','pass','pass']
```

```python
encoded_result=label_encoder.fit_transform(exam_result)
print(encoded_result)  # 1 for pass and 0 for fail transformed
```

```python
data=pd.DataFrame({'study_time':study_time,'marks_scored':marks_scored,'exam_result':encoded_result})
```

```python
y=data.drop('exam_result',axis=1)
#y=pd.DataFrame({'study_time':study_time,'marks_scored':marks_scored})
# both the ways are same to remove the exam_result column
x=data['exam_result']
print(x,y)
```

**USING MLP CLASSIFIER **

```python
clf=MLPClassifier(hidden_layer_sizes=(4,),activation='logistic',max_iter=1000,random_state=42)
clf.fit(y,x)
```

```python
stu_data=pd.DataFrame({'study_time':[2.5],'marks_scored':[70]}) #hrs studied & previous exam score
predicted_result=clf.predict(stu_data)
# print(predicted_result)
```

```python
# decodeing is done in the answer  manual way:-
# if predicted_result[0]==1:
#     print('pass')
# else:
#     print('fail')

# by module
result_decode=label_encoder.inverse_transform(predicted_result)
print(result_decode)
```
