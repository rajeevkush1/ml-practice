# REQUESTED DATA WITH THE HELP OF AN **API**

```python
import requests

url = "https://real-time-amazon-data.p.rapidapi.com/seller-products"

querystring = {"seller_id":"A02211013Q5HP3OMSZC7W","country":"US","page":"1","sort_by":"RELEVANCE"}

headers = {
	"x-rapidapi-key": "3556b037e2msh5e756f8cae10019p156937jsnf2affca428f4",
	"x-rapidapi-host": "real-time-amazon-data.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

response.json()['parameters']
```

EXTRACTED DATA FROM 1 PAGE OF AMAZON

```python
pd.DataFrame(response.json()['data']['seller_products'])
```

```python
df=pd.DataFrame()
df
```
