```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=np.array([[30,50000,1],[25,90000,3],[35,20000,0],[40,45000,2],[50,60000,1],[20,95000,1]])
labels = np.array([1,1,0,1,2,2]) #0=low ,1=medium ,3=high
df=pd.DataFrame(data,columns=['Age','Salary','Purchased'])
df
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,labels,test_size=0.2,random_state=42)
```

```python
scalar= StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
knn=KNeighborsClassifier(n_neighbors=3)  #no of neighbours =3(n_neighbors)
knn.fit(x_train,y_train)
```

KNN ALREADY HAVE ACCURACY MODULE WE DONT HAVE TO IMPORT IT

```python
accuracy=knn.score(x_test,y_test)
print(f"model_accuracy :{accuracy}")
```
