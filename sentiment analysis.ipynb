```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Load Dataset
file_path = "/content/sentimentdataset.csv"
df = pd.read_csv(file_path)

df = df[['Text', 'Sentiment']]  # Keep only relevant columns

# Encode Sentiment Labels
label_encoder = LabelEncoder()
df['Sentiment'] = label_encoder.fit_transform(df['Sentiment'])

# Custom Stopwords List
custom_stopwords = set([
    "the", "is", "in", "and", "to", "of", "it", "for", "on", "this", "with", "as", "that", "at",
    "from", "by", "an", "be", "are", "was", "were", "has", "have", "had", "but", "not", "or", "so"
])

# Text Cleaning Function
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\W', ' ', text)  # Remove special characters
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    words = text.split()
    words = [word for word in words if word not in custom_stopwords]
    return ' '.join(words)

# Apply Text Cleaning
df['Cleaned_Text'] = df['Text'].apply(clean_text)

# Split Data
X_train, X_test, y_train, y_test = train_test_split(df['Cleaned_Text'], df['Sentiment'], test_size=0.2, random_state=42)

# Apply TF-IDF Vectorization
vectorizer = TfidfVectorizer(max_features=7000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Define Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(n_estimators=100),
    "SVM": SVC(kernel='linear'),
    "NaÃ¯ve Bayes": MultinomialNB()
}

# Train & Evaluate Models
results = {}
for name, model in models.items():
    model.fit(X_train_tfidf, y_train)
    y_pred = model.predict(X_test_tfidf)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy

# Sort and Print Accuracy
sorted_results = dict(sorted(results.items(), key=lambda item: item[1], reverse=True))
print("Model Performance (Accuracy %):")
for model, acc in sorted_results.items():
    print(f"{model}: {acc * 100:.2f}%")
```
